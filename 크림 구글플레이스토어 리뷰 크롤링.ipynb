```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
# 크롤링 리뷰페이지 URL 입력
url = 'https://play.google.com/store/apps/details?id=com.fstudio.kream'
```

```python
driver = webdriver.Chrome("chromedriver")
driver.get(url)
time.sleep(3)
```

```python
# 리뷰 전체 보기 버튼 클릭
driver.find_element(By.XPATH,'//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div[2]/div/div[1]/c-wiz[4]/section/div/div/div[5]/div/div/button/span').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
content = soup.findAll("div", {'class' : 'RHo1pe'})
print(len(content))
```

```python
item_list = [ 
"iXRFPc", #별점 
"h3YV2d", #리뷰
"AJTPZc" #리뷰 공감수 
]

# 수집 대상 요소를 컬럼으로 하는 빈 list 생성
tmp_list = []
print(len(content))
for i in range(len(content)):
    try:
        item1 = content[i].find("div", {'class' : item_list[1]}).text
    except:
        itme1 = ""
    try:
        item2 = content[i].find("div", {'class': item_list[0]})['aria-label']
    except:
        item2 = ""
    try:
        item3 = content[i].find("div", {'class' : item_list[2]}).text
    except:
        item3 = str(0)
    tmp_list.append([item1, item2, item3])
    
tmp_list
```

```python
import pandas as pd
from pandas import DataFrame
from datetime import datetime

result_df = DataFrame(tmp_list, columns = ['리뷰내용', '별점', '리뷰공감수'])

result_df.head()
```

```python
tmp = result_df.copy()

tmp['별점'] = tmp['별점'].apply(lambda x: x[5:])

m = re.compile('[0-9][\.0-9]*')
tmp['별점'] = tmp['별점'].apply(lambda x : m.findall(x)[0])
tmp['리뷰공감수'] = tmp['리뷰공감수'].apply(lambda x : m.findall(x)[0])
tmp.head(3)
```

```python
tmp.to_csv('크림.csv')
```
